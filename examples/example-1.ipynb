```julia
using Plots
using Images
using StatsBase
```

```julia
using SimradEK60
```

Load all the pings from our sample krill swarm, courtesy British Antarctic Survey

```julia
filename = joinpath(dirname(@__FILE__),
                       "../test/data/krill_swarm_20091215/JR230-D20091215-T121917.raw");
```

```julia
ps = collect(pings(filename));
```

```julia
ps38 = [p for p in ps if p.frequency ==38000];
ps120 = [p for p in ps if p.frequency ==120000];
ps200 = [p for p in ps if p.frequency ==200000];
Sv38 = Sv(ps38); # 38 kHz volume backscatter
Sv120 = Sv(ps120); # 120 kHz
Sv200 = Sv(ps200); # 200 kHz
_R = R(ps38); # Range
```

# Visualise echograms as a grayscale images

```julia
function grayscale(a::Matrix)
    a = Float64.(a)
    b = minimum(a)
    c = maximum(a)
    d = (a .- b) / (c -b)
    colorview(Gray, d)
end
```

```julia
imresize(grayscale(Sv38),(300,400))
```

## Zoom in

```julia
r = imresize(grayscale(Sv38[300:1100,410:520]),(300,400))
```

```julia
g = imresize(grayscale(Sv120[300:1100,410:520]),(300,400))
```

```julia
b = imresize(grayscale(Sv200[300:1100,410:520]),(300,400))
```

Make a false colour composite

```julia
colorview(RGB,r,g,b)
```

```julia
const EK500 = cgrad([RGB(1, 1, 1),
           RGB(159/255, 159/255, 159/255),
           RGB(95/255, 95/255, 95/255),
           RGB(0, 0, 1),
           RGB(0, 0, 127/255),
           RGB(0, 191/255, 0),
           RGB(0, 127/255, 0),
           RGB(1, 1, 0),
           RGB(1, 127/255, 0),
           RGB(1, 0, 191/255),
           RGB(1, 0, 0),
           RGB(166/255, 83/255, 60/255),
           RGB(120/255, 60/255, 40/255)])

function myechogram(a; vmin = nothing, vmax = nothing, leg=true, color=EK500, size=(640, 480))
    if vmax != nothing
        a[a .> vmax] = vmax
    end

    if vmin != nothing
        a[a .< vmin] = vmin
    end

    p = plot(layout=1, leg=leg, size=size)
    Plots.heatmap!(a, yflip=true, color=color)
    p
end
```

```julia
p = myechogram(Sv200[300:1100,410:520], vmin = -95, vmax=50)
title!("Example krill swarm")
xlabel!("Pixels")
ylabel!("Pixels")
```
